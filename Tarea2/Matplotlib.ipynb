## Insertar figuras en otras. Axes
Son parecidos a los subplots, pero estos permiten colocar figuras dentro de otra figura con los mismos ejes.

```python
x = np.linspace(4, 10, 70)
y = (x+7) ** 5

fig = plt.figure() #Crear una figura nueva o activar una figura existente

# fig.add_axes: ejes nuevos [izquierda, abajo, ancho, alto]
axes1 = fig.add_axes([4.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([4.2, 0.6, 0.3, 0.2]) # inset axes

# Figura grande
axes1.plot(x+2, y, 'y')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Imagen grande')

# Figura insertada
axes2.plot(y, x+5, 'r')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Imagen pequeña');
```

## Títulos, etiquetas y leyendas
Los títulos (titles) son el nombre que tienen los gráficos, las etiquetas (labels) son los nombres de los ejes y las leyendas (legends) son para guiarse cuando tenemos más de una trazado en el gráfico y se quieren diferenciarlas.

```python
# Titulo y etiquetas
x = np.linspace(0, 10, 100)
y=np.sin(x)  # Función de numpy que me da el seno de x
 
plt.plot(x, y)  # argumentos simples para el scatter plot.
plt.xlabel('X')  # etiqueta del eje x
plt.ylabel('Sen(X)')  # etiqueta del eje y
 
plt.title('Seno entre 0 y 10')  # título del gráfico
plt.show() # Para que muestre el gráfico
```

```python
# También se puede insertar código Latex
x = np.linspace(0, 10, 100)
y=np.sin(x)  # Función de numpy que me da el seno de x
 
plt.plot(x, y, color = "red")  # argumentos simples para el scatter plot.
plt.xlabel(r"$\theta$")  # etiqueta del eje x
plt.ylabel(r'$Sen(\theta)$')  # etiqueta del eje y
 
plt.title(r"Valores para $Sen(\theta)$")  # título del gráfico
plt.show() # Para que muestre el gráfico
```

## Gráficos de barras
Estos sirven para analizar una variable categórica.

```python
Estratos = [1,2,3,4,5,6] #posibles valores
y = [23,54,32,17,9,4] # altura de cada barra. En este caso, la cantidad de personas en cada estrato
 
plt.bar(Estratos, y, label='Estratos') # con plt.bar creo el barplot.
 
plt.xlabel('Estratos')
plt.ylabel('Total')
 
plt.title('Conteo de personas en cada estrato')
plt.legend()
plt.show()
```

```python
ingresos = ['Bajo', 'Medio', 'Alto']
casados = [5,20,15]
solteros = [40,5,38]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, casados, width, label='Casados')
rects2 = ax.bar(x + width/2, solteros, width, label='Solteros')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Total personas')
ax.set_title('Total de personas por su nivel de ingreso disgregado por su estado civil')
ax.set_xticks(x)
ax.set_xticklabels(ingresos)
ax.legend()

### Esto es para que me dé el número dentro de cada categoría 
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

Se observa que para las personas de ingreso bajo y alto predominan las personas solteras, mientras que para las personas de ingreso medio, la mayoría está casada.

## Boxplots

```python
# Fijar semilla
np.random.seed(1004)

# Generación de datos aleatorios
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low)) # concatena los valores creados anteriormente

fig, axs = plt.subplots(2, 3) # 2 filas y 3 columnas

# Box-plot básico
axs[0, 0].boxplot(data)
axs[0, 0].set_title('Box-plot básico')

# notched plot
axs[0, 1].boxplot(data, 1)
axs[0, 1].set_title('Plot notched')

# Cambiar carateristicas outliers
axs[0, 2].boxplot(data, 0, 'rD') #r=red
axs[0, 2].set_title('Cambiar carateristicas outliers')

# No mostrar outliers
axs[1, 0].boxplot(data, 0, '')
axs[1, 0].set_title("No mostrar outliers")

# Cajas horizontales
axs[1, 1].boxplot(data, 0, 'bs', 0)
axs[1, 1].set_title('Cajas horizontales')

# Cambiar tamaño bigotes
axs[1, 2].boxplot(data, 0, 'bs', 0, 0.75)
axs[1, 2].set_title('change whisker length')

#Ajuste de los subplots
fig.subplots_adjust(left=0.08, right=2, bottom=0.05, top=0.9,
                    hspace=0.4, wspace=0.3)
plt.show()
```

## Densidades

```python
data = [1,1,1,3,3,3,3,4,5,6,7,8,8,8,8,9,5,55,6,9,8,7,23,31]
density = kde.gaussian_kde(data) #Método:  kde = kernel density
x = np.linspace(-2,10,300)
y=density(x)

plt.plot(x,y)
plt.title("Density Plot of the data")
plt.show()
```

no se muestran todos los valores. Hace falta establecer el límite de x

```python
data = [1,1,1,3,3,3,3,4,5,6,7,8,8,8,8,9,5,55,6,9,8,7,23,31,10]
prob_density = kde.gaussian_kde(data)
prob_density.covariance_factor = lambda : .25
prob_density._compute_covariance()

x = np.linspace(-2,10,300)
y=prob_density(x)

plt.plot(x,y)
plt.title("Density Plot of the data")
plt.show()
```

```python
#Primer grupo de datos aleatorios generados
np.random.seed(1004)

spread = np.random.rand(30) * 70
center = np.ones(25) * 50
flier_high = np.random.rand(30) * 70 + 100
flier_low = np.random.rand(10) * -10
data = np.concatenate((spread, center, flier_high, flier_low))


# Más realizaciones de datos generados
spread = np.random.rand(50) * 100
center = np.ones(25) * 40
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
d2 = np.concatenate((spread, center, flier_high, flier_low))
data = [data, d2, d2[::3]]

# Multiples box-plot con mismos ejes
fig, ax = plt.subplots()
ax.boxplot(data, 0, 'b')
plt.title("Múltiples boxplots con el mismo eje")
plt.show()
```
